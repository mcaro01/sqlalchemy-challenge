# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine(f"sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement= Base.classes.measurement
station= Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
engine.execute("SELECT * FROM measurement").all()
```

```python
#first.row = session.query(measurement).first()
#first.row.__dict__
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
recent_date = session.query(measurement.date).order_by(measurement.date.desc()).first().date
recent_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
recent_data_point = session.query(measurement.date, measurement.prcp).order_by(func.date(measurement.date).desc()).all()
recent_data_point
```

```python
# Calculate the date one year from the last date in data set.

year_ago = dt.date(2017,8,23)- dt.timedelta(days=365)
print(year_ago)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
recent_data_point = session.query(measurement.date, measurement.prcp).order_by(func.date(measurement.date).desc()).first()
recent_data_point
```

```python
# Perform a query to retrieve the data and precipitation scores
precp_results = session.query(measurement.id, measurement.station, measurement.date, measurement.prcp, measurement.tobs).\
                    filter(measurement.date >= year_ago).all()
precp_results
```

```python
# Save the query results as a Pandas DataFrame. Explicitly set the column names
df_old = pd.DataFrame(precp_results, columns = ["ID", "Station", "Date", "Precipitation", "TOBS"])
df_old
```

```python
# Sort the dataframe by date
sort_date = df_old.sort_values(by='Date', ascending = True)
sort_date
```

```python
#sort_date.plot.bar()
sort_date.plot(x = "Date", y = "Precipitation")
plt.title("Precipitation")
plt.ylabel("Inches")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
countp =df_old["Precipitation"].count()
meanp = df_old["Precipitation"].mean()
stdp = df_old["Precipitation"].std()
minp = df_old["Precipitation"].min()
twentyfive = df_old["Precipitation"].quantile(.25)
fifty = df_old["Precipitation"].quantile(.50)
seventyfive =df_old["Precipitation"].quantile(.75)
maxp = df_old["Precipitation"].max()

list1 = [countp, meanp, stdp, minp, twentyfive, fifty, seventyfive, maxp]

summary_stats = pd.DataFrame(list1, index = ["count", "mean", "std", "min", "25%", "50%", "75%", "max"])
summary_stats.columns = ["precipitation"]
summary_stats
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
engine.execute('SELECT COUNT(DISTINCT station) FROM measurement').all()
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)

session.query(measurement.station).distinct().count()
```

```python
# List the stations and their counts in descending order.
most_active_station_id = session.query(measurement.station,func.count(measurement.station)).\
                               group_by(measurement.station).\
                               order_by(func.count(measurement.station).desc()).all()
most_active_station_id
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
temp_calculation = session.query(func.min(measurement.tobs),func.max(measurement.tobs),func.avg(measurement.tobs)).\
                filter(measurement.station == 'USC00519281').order_by(func.min(measurement.tobs)).all()
temp_calculation 
```

```python
temp_calculation = session.query(func.min(measurement.tobs),func.max(measurement.tobs),func.avg(measurement.tobs)).\
                filter(measurement.station == 'USC00519281').all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

year_temp = session.query(measurement.tobs, measurement.station).\
      filter(measurement.date >= year_ago, measurement.station == 'USC00519281').\
      order_by(measurement.tobs).all()
df1 = pd.DataFrame(year_temp)
#df1 = pd.DataFrame(year_temp), columns=['Tobs'])
df1.set_index('station', inplace=True)
df1.head(100)

```

```python
#plot the results as a histogram
bins = 12
df1.hist(by='station', bins=12, figsize=(15,9),)
plt.grid()
plt.xlabel("Temperature", fontsize=16)
plt.ylabel("Frequency")
plt.grid()
plt.title('tobs')
plt.tight_layout()
# Show plot
plt.show()
```
